```python
import pandas as pd
```

```python
from src import DATA_FOLDER
```

```python
import os
```

```python
files = sorted([os.path.join(DATA_FOLDER, file) for file in os.listdir(DATA_FOLDER) if ".csv.gz" in file])
files
```

```python
wave_files = list(filter(lambda n: "wave" in n, files))
wave_files
```

```python
dfs = (pd.read_csv(f) for f in wave_files)
df_wave   = pd.concat(dfs, ignore_index=True)
df_wave = df_wave.loc[:, ~df_wave.columns.str.contains('^Unnamed')]
df_wave.drop_duplicates(keep="last", inplace=True)
df_wave.timestamp = pd.to_datetime(df_wave.timestamp, utc=True)
df_wave.head()
```
