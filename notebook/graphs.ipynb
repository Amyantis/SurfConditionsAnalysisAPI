```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.models.sources import ColumnDataSource
from bokeh.models import HoverTool
```

```python
source = ColumnDataSource(df_wave)
p = figure(x_axis_type="datetime", height=150)
p.line(x='timestamp', y='surf.max', color='darkmagenta', source=source, legend="surf max")
p.line(x='timestamp', y='surf.min', color='mediumblue', source=source, legend="surf min")
show(p)
```

```python
source = ColumnDataSource(df_tides)
p = figure(x_axis_type="datetime", height=150)
p.line(x='timestamp', y='height', color='darkcyan', source=source, legend="tides height")
show(p)
```

```python
source = ColumnDataSource(df_weather)
p = figure(x_axis_type="datetime", height=150)
p.line(x='timestamp', y='temperature', color='tomato', source=source, legend="temperature(Â°C)")
show(p)
```

```python
source = ColumnDataSource(df_wind)
p = figure(x_axis_type="datetime", height=150)
p.line(x='timestamp', y='speed', color='peru', source=source, legend="wind speed (Km/H)")
show(p)
```
