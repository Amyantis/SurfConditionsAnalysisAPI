```python
URL = "https://services.surfline.com/kbyg/mapview?south=-90&west=-180&north=90&east=180"
```

```python
def get_data():
    with urlopen(URL) as fdesc:
        data = fdesc.read()
        return json.loads(data.decode("utf8"))
```

```python
data = get_data()
```

```python
pd.DataFrame(data)
```

```python
df_spots = pd.io.json.json_normalize(data["data"]["spots"])
df_spots.head()
```
