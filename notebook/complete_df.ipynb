```python
df_weather = get_weather_df()
df_weather.head()
```

```python
df_wind = get_wind_df()
df_wind.head()
```

```python
import pandas as pd
```

```python
def suffix(df, suffix="", inplace=False):
    return df.rename(columns={c: suffix + c for c in df.columns if c != "timestamp"}, inplace=inplace)
```

```python
df = pd.merge(suffix(df_wave, "wave_"), suffix(df_wind, "wind_"), on='timestamp')
df = pd.merge(df, suffix(df_weather, "weather_"), on='timestamp')
df = pd.merge(df, suffix(df_tides, "tides_"), on='timestamp')
df.columns
```

```python
from datetime import datetime
import pytz

df = df[df.timestamp > datetime.utcnow().replace(tzinfo=pytz.utc)]
df
```
